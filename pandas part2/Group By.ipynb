```python
import numpy as np
import pandas as pd

df=pd.read_csv('student_completed_data.csv')
df.loc[df['FullName']=='Afsana Mimi','Total Marks']=np.nan
df.loc[df['FullName']=='Sadia Chowdhury','Total Marks']=np.nan
df.loc[df['FullName']=='Alif Rahman','Total Marks']=np.nan

df
```

```python
instructor_group=df.groupby('Instructor')

for instructor,df_ins in instructor_group:
    print(instructor)
    display(df_ins)
```

```python
instructor_group.describe()
```

```python
mean_mark=instructor_group['Total Marks'].mean()
mean_mark
```

```python
# fill nan with mean value

df['Total Marks']=df['Total Marks'].fillna(df['Instructor'].map(mean_mark))
df
```
