```python
def check(x):
    if x == 4.00:
        return "A+"
    elif x >= 3.75:
        return "A"
    elif x >= 3.60:
        return "A-"
    elif x >= 3.50:
        return "B+"
    elif x >= 3.25:
        return "B"
    elif x >= 3.00:
        return "B-"
    elif x >= 2.75:
        return "C+"
    elif x >= 2.50:
        return "C"
    elif x >= 2.25:
        return "D"
    else:
        return "F"

df.loc[0,'CGPA']=4.00
df['result']=df['CGPA'].apply(check)

df
```
