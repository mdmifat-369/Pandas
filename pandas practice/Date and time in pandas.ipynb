```python
import pandas as pd

data = {
    "date": ["2025-01-01", "2025-02-15", "2025-03-10"]
}

df=pd.DataFrame(data)

df['date']=pd.to_datetime(df['date'])
```

```python
#🧠 2. Extract parts (.dt)

df['year']=df['date'].dt.year
df['month']=df['date'].dt.month
df['day']=df['date'].dt.day
df['week day']=df['date'].dt.day_name()
df['month name']=df['date'].dt.month_name()

df
```

```python
#⏳ 3. Time difference
import pandas as pd

df = pd.DataFrame({
    "start": ["2025-01-01", "2025-01-10"],
    "end": ["2025-01-05", "2025-01-20"]
})

df['start']=pd.to_datetime(df['start'])
df['end']=pd.to_datetime(df['end'])

df['diff']=(df['end']-df['start'])

df
```

```python
#shift key word

import pandas as pd

df = pd.DataFrame({
    "date": ["2025-01-01", "2025-01-21", "2025-02-01"]
})

df['date']=pd.to_datetime(df['date'])
df.sort_values('date')

df['prev_date']=df['date'].shift(1)

df['gap_day']=df['date']-df['prev_date']

df
```

```python
import pandas as pd

df = pd.DataFrame({
    "start": ["2025-01-01 10:00:00"],
    "end": ["2025-01-03 12:00:00"]
})

df['start'] = pd.to_datetime(df['start'])
df['end'] = pd.to_datetime(df['end'])

df['diff'] = df['end'] - df['start']

df['days'] = df['diff'].dt.days
df['hours'] = df['diff'].dt.total_seconds() / 3600

df
```

```python
#⏩ 6. Add / subtract time
import pandas as pd

df = pd.DataFrame({
    "date": ["2025-01-01", "2025-01-10"]
})

df['date']=pd.to_datetime(df['date'])

df['next_day']=df['date']+pd.Timedelta(days=35)
df['prev_day']=df['date']-pd.Timedelta(days=9)

df
```
