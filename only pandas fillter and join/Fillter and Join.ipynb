```python
import pandas as pd

# df1: Employees list
df1 = pd.DataFrame({
    'emp_id': [1, 2, 3],
    'name': ['Rahim', 'Karim', 'Sami']
})

# df2: Salary list
df2 = pd.DataFrame({
    'emp_id': [2, 3, 4],
    'salary': [50000, 60000, 70000]
})
```

```python
result = pd.merge(df1, df2, on='emp_id', how='inner')
print(result)
```

```python
result = pd.merge(df1, df2, on='emp_id', how='left')
print(result)
```

```python
result = pd.merge(df1, df2, on='emp_id', how='right')
print(result)
```

```python
result = pd.merge(df1, df2, on='emp_id', how='outer')
print(result)
```
