```python
import numpy as np
import pandas as pd
```


                                                     Creating a DataFrame

                                                     

```python
# when u use dictionary you can set columns name as a key then u do not set columns name manually
data={
    'Name':['John','Anna','Peter','Linda'],
    'Age':[28,34,29,42],
    'City':['New York','Paris','Berlin','London'],
    'Salary':[65000,70000,62000,85000]
}

pd.DataFrame(data)
```

```python
# when u use list then u need to set columns name manually example
data_list=[
    ['John',28,'New York',65000],
    ['Anna',34,'Paris',70000],
    ['peter',29,'Berlin',62000],
    ['Linda',42,'London',85000]
]

df1=pd.DataFrame(data_list)
df1
```

```python
#set manually
col=["name","Age","City","Salary"]
df2=pd.DataFrame(data_list,columns=col)
df2
```


                                           Selection and indexing of Columns


```python
data_list=[
    ['John',28,'New York',65000],
    ['Anna',34,'Paris',70000],
    ['peter',29,'Berlin',62000],
    ['Linda',42,'London',85000]
]

col=["Name","Age","City","Salary"]
df2=pd.DataFrame(data_list,columns=col)
df2
```

```python
# access a single columns
df2['Name']
```

```python
# access a multiple colums at a time
df2[['Name','City','Age']]
```

```python
# Add a new columns in the data Frame
df2["Designation"]=["Doctor","Student","Teacher","Engineer"]
df2
```


                                             Remove columns and Rows
          

```python
# if you do not use inplace key word then it do not remove permanently
df2.drop('Designation',axis=1)
```

```python
# this time use implace 
df2.drop('Designation',axis=1,inplace=True)
df2
```

```python
# if you want remove multiple colmuns
df2.drop(['Name','City'],axis=1)
```

```python
# if you want remove Row
df2.drop(0,axis=0)
```
