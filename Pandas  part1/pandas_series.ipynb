Pandas Series

A Series is a one-dimensional labeled array capable of holding any data type. The axis labels are collecyively called the index.

Creating Series 
There are multiple ways to create a pandas Series:

```python
import numpy as np
import pandas as pd
```

```python
lables=['a','b','c']
mylist=[1,2,3]
a=np.array([1,2,3])
d={1:10,2:20,3:30}
```

```python
pd.Series(mylist)
```

```python
pd.Series(mylist,index=lables)
```

```python
pd.Series(a,index=mylist)
```

```python
pd.Series(d)
```
