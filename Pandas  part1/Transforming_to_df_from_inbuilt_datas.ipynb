```python
import pandas as pd
my_list=[['Alice',25],['Bob',30],['Charile',28]]
list_df=pd.DataFrame(my_list,columns=['Name','age'],index=['a','b','c'])

list_df
```

```python
my_tuple=(('Alice',25),('Bob',30),('Charlie',28))
tuple_df=pd.DataFrame(my_tuple,columns=['Name','age'],index=['a','b','c'])

print(tuple_df)
```

```python
my_dict={
    'Name':['Alice','Bob','Charlie'],
    'Age':[25,30,28]
}

my_dict_df=pd.DataFrame(my_dict)
print(my_dict_df)
```
